#**NumPy - Numerical Computing**

**NumPy** is a powerful library for numerical computations in Python. It provides support for multi-dimensional arrays and mathematical functions.

### Key Features:
- **Arrays**: Store data in a grid-like structure.
- **Vectorized Operations**: Perform operations on entire arrays without loops.
- **Random Number Generation**: Create random data for simulations.


```python
import numpy as np
```

```python
# 1. Create a NumPy array
array = np.array([1, 2, 3, 4, 5])
print("Array:", array)
```

```python
# 2. Create a 2D array (matrix)
matrix = np.array([[1, 2], [3, 4], [5, 6]])
print("Matrix:\n", matrix)
```

```python
# 3. Generate an array of zeros
zeros = np.zeros((2, 3))
print("Zeros Array:\n", zeros)
```

```python
# 4. Perform element-wise operations
squared = array ** 2
print("Squared Elements:", squared)
```

```python
# 5. Calculate statistics
mean = np.mean(array)
print("Mean:", mean)
```

```python
# 6. Generate random numbers
random_numbers = np.random.rand(5)
print("Random Numbers:", random_numbers)
```

# **Pandas - Data Manipulation and Analysis**

**Pandas** is a library for working with structured data like tables (DataFrames) and series. It simplifies data manipulation tasks like cleaning, transforming, and aggregating.

### Key Features:
- **DataFrames**: Tabular data with labeled rows and columns.
- **Series**: One-dimensional labeled arrays.
- **Data Cleaning**: Handle missing values, duplicates, etc.
- **Group Operations**: Summarize data by groups.



```python
import pandas as pd
```

```python
# 1. Create a Series
ages = pd.Series([25, 30, 35], index=['Alice', 'Bob', 'Charlie'])
print("Series:\n", ages)

# Access elements in a Series
print("Access Bob's Age:", ages['Bob'])

# Perform operations on Series
ages += 5
print("Updated Ages:\n", ages)
```

```python
# 2. Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Score': [90, 80, 85]
}
df = pd.DataFrame(data)
print("\nDataFrame:\n", df)
```

```python
# 3. Access a column
print("\nNames:\n", df['Name'])
```

```python
# 4. Add a new column
df['Passed'] = df['Score'] > 85
print("Updated DataFrame:\n", df)
```

```python
# 5. Handle missing values
df.loc[3] = ['Diana', None, None, None]  # Add a row with missing values
df['Age'] = df['Age'].fillna(0)  # Fill missing ages with 0
print("DataFrame with Missing Values Handled:\n", df)
```

```python
# 6. Group data
grouped = df.groupby('Passed')['Score'].mean()
print("Average Score by Passed Status:\n", grouped)
```

**Matplotlib** is a library for creating static, animated, and interactive visualizations.

### Key Features:
- **Line Charts**: For trends over time.
- **Bar Charts**: For categorical data.
- **Scatter Plots**: For relationships between two variables.
- **Histograms**: To visualize distributions.
- **Pie Charts**: For proportional data.
- **Stacked Bar Charts**: To show parts of a whole.


```python
import matplotlib.pyplot as plt

# 1. Line Plot
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y, label='Prime Numbers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')
plt.legend()
plt.show()
```

```python
# 2. Bar Plot
categories = ['A', 'B', 'C']
values = [10, 15, 7]
plt.bar(categories, values, color='skyblue')
plt.title('Bar Plot Example')
plt.show()
```

```python
# 3. Scatter Plot
plt.scatter(x, y, color='red')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# 4. Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5, 5]
plt.hist(data, bins=5, color='purple', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

```python
# 5. Pie Chart
labels = ['A', 'B', 'C']
sizes = [30, 50, 20]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'lightcoral'])
plt.title('Pie Chart Example')
plt.axis('equal')  # Equal aspect ratio to make the pie circular
plt.show()
```

```python
# 6. Stacked Bar Chart
categories = ['A', 'B', 'C']
group1 = [5, 7, 6]
group2 = [3, 8, 2]
bar_width = 0.6

# Create stacked bars
plt.bar(categories, group1, label='Group 1', color='lightgreen')
plt.bar(categories, group2, bottom=group1, label='Group 2', color='lightblue')
plt.title('Stacked Bar Chart Example')
plt.legend()
plt.show()
```

# **Seaborn - Statistical Data Visualization**

**Seaborn** is a library built on Matplotlib for creating attractive and informative statistical graphics.

### Key Features:
- **Heatmaps**: Display correlations or tabular data.
- **Boxplots**: Visualize data distributions.
- **Pairplots**: Explore relationships in a dataset.
- **Violin Plots**: Combine boxplots and KDEs.
- **Count Plots**: Display frequency of categories.
- **Regression Plots**: Show trends and relationships.
- **Bar Plots**: Summarize data for categories.
.


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load a sample dataset
tips = sns.load_dataset('tips')
```

```python

# 1. Heatmap
numerical_columns = tips.select_dtypes(include=['float64', 'int64'])
corr = numerical_columns.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Example')
plt.show()

```

```python

# 2. Boxplot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Boxplot Example')
plt.show()
```

```python
# 3. Pairplot
sns.pairplot(tips, hue='sex')
plt.suptitle('Pairplot Example', y=1.02)
plt.show()
```

```python
# 4. Violin Plot
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, split=True)
plt.title('Violin Plot Example')
plt.show()
```

```python
# 5. Count Plot
sns.countplot(x='day', hue='sex', data=tips, palette='pastel')
plt.title('Count Plot Example')
plt.show()
```

```python
# 6. Regression Plot
sns.regplot(x='total_bill', y='tip', data=tips, line_kws={"color": "red"})
plt.title('Regression Plot Example')
plt.show()
```

```python
# 7. Bar Plot
sns.barplot(x='day', y='total_bill', hue='sex', data=tips, errorbar=None, palette='muted')
plt.title('Bar Plot Example')
plt.show()
```
